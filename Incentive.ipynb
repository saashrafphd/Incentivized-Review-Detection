```python
import numpy as np
temp= b.groupby(['prod_id'])['CompSent'].agg(np.mean).reset_index()
temp.rename(columns={'CompSent':'avg_PSent'}, inplace=True)
b = pd.merge(b, temp, how='outer', on=['prod_id'])
temp= b.groupby(['prod_id'])['CompSent'].agg(np.median).reset_index()
temp.rename(columns={'CompSent':'med_PSent'}, inplace=True)
b = pd.merge(b, temp, how='outer', on=['prod_id'])
```

```python
b['fSent'] = ""
for prod in b['prod_id'].unique():
    dftemp = b.loc[(b['prod_id'] == prod)]
    stdavg = ((sum((dftemp['CompSent']-dftemp['avg_PSent'])**2))/dftemp.shape[0])**(1/2)
    stdmed = ((sum((dftemp['CompSent']-dftemp['med_PSent'])**2))/dftemp.shape[0])**(1/2)
    if(stdavg<stdmed):
        b.loc[b.prod_id == prod, "fSent"] = dftemp['avg_PSent'].iloc[0]
    else:
        b.loc[b.prod_id == prod, "fSent"] = dftemp['med_PSent'].iloc[0]
b['inc'] = b.apply(lambda x: 1 if ((x['CompSent']-x['fSent'])**2 > .6) else 0, axis=1) 
b['incentivised'] = b.apply(lambda x: 1 if (x['burst'] & x['inc']) else 0, axis=1) 
b
```
